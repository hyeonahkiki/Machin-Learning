# ridge regression(릿지회귀)

```python
from sklearn.linear_model import Ridge
# data_set의 split과정아 없으면 X_train이 정의되지 않았다고 에러가 발생

# 보스턴 데이터 셋 들고오기 /  train_test_split 가져오기
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

boston = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston['data'], boston['target'])
ridge = Ridge().fit(X_train, y_train)
```

```python
# 훈련세트점수, 테스트세트 점수
print(ridge.score(X_train, y_train))
print(ridge.score(X_test, y_test))
```

# linear_regression(선형회귀)

```python
# 선형회귀
from sklearn.linear_model import LinearRegression

boston = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston['data'], boston['target'])
linear_regression = LinearRegression().fit(X_train, y_train)
```

```python
print(linear_regression.score(X_train, y_train))
print(linear_regression.score(X_test, y_test))
```

# ridge regression(alpha 사용)

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print(ridge10.score(X_train, y_train))
print(ridge10.score(X_test, y_test))
```

```python
ridge1 = Ridge(alpha=1).fit(X_train, y_train)
print(ridge1.score(X_train, y_train))
print(ridge1.score(X_test, y_test))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print(ridge01.score(X_train, y_train))
print(ridge01.score(X_test, y_test))
```
